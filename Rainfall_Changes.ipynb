```python
#import library
import ee
import geemap
```

```python
#Authenticate and Intialize
ee.Authenticate()
```

```python
ee.Initialize()
```

```python
#Load the map
Map=geemap.Map()
Map
```

```python
# Define the Earth Engine dataset (e.g., CHIRPS precipitation data)
chirps = ee.ImageCollection("UCSB-CHG/CHIRPS/PENTAD")

# Filter the collection by date
start_date = '2020-01-01'
end_date = '2021-01-01'
filtered = chirps.filterDate(start_date, end_date)

total = filtered.sum()

#Define Visualization Parameters
palette = ['white', 'blue']
vis_params = {
    'min': 0,
    'max': 2000,
    'palette': palette
}

# Add the layer to the Map
Map.addLayer(total, vis_params, 'Total Precipitation')
```
